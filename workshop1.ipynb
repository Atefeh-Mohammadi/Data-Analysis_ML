```python
#regression
```

```python
import pandas as pd
import numpy as np
```

```python
X = df[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']]
y = df['salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
predictions = lm.predict(x_test)
```

```python
sns.distplot((y_test - predictions),bins=30)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)
```

```python
# decision tree
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['degree_p', 'etest_p', 'mba_p', 'salary']]
y = df['status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(accuracy_score(y_test, predictions))
```

```python
# random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
print(confusion_matrix(y_test, rfc_pred))
```

```python
print(accuracy_score(y_test, rfc_pred))
```

```python
# k means clustering
```

```python
df["status"] = df["status"].astype('category')
df.dtypes
```

```python
df["binary_status"] = df["status"].cat.codes
df.head()
```

```python
sns.lmplot('etest_p','salary',data=df, hue='status',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df[['etest_p', 'salary']])
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(df['etest_p'],df['salary'],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(df['etest_p'],df['salary'],c=df['binary_status'],cmap='rainbow')
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(df[['etest_p', 'salary']])
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(df['etest_p'],df['salary'],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(df['etest_p'],df['salary'],c=df['binary_status'],cmap='rainbow')
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(df[['etest_p', 'salary']])
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(df['etest_p'],df['salary'],c=kmeans.labels_,cmap='coolwarm')
ax2.set_title("Original")
ax2.scatter(df['etest_p'],df['salary'],c=df['binary_status'],cmap='summer')
```

```python
print(confusion_matrix(df['binary_status'],kmeans.labels_))
print()
print(classification_report(df['binary_status'],kmeans.labels_))
```
